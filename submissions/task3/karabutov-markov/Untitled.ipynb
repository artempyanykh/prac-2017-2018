```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import csv
```

```python
def conc(ar1, ar2):
    for i in range(len(ar2)):
        ar1.append(ar2[i])
    return ar1

def sum_arr(ar1, ar2):
    res = []
    for i in range(len(ar1)):
        s = ar1[i][2] + ar2[i][2]
        s1 = ar1[i][3] + ar2[i][3]
        s2 = ar1[i][4] + ar2[i][4]
        s3 = ar1[i][5] + ar2[i][5]
        res.append([ar1[i][0], ar1[i][1], s, s1, s2, s3])
    return res

def make_csv(data, name):
    with open(name, 'w') as fp:
        writer = csv.writer(fp, delimiter=',')
        writer.writerows(data)


def gen_names():
    s = ["b", "m", "s"]
    n = [2, 2, 5]
    names = []
    for i in range(3):
        for j in range(n[i]):
            s1 = "MS-" + s[i] + str(j+1) + "-" + "inventory.csv"
            s2 = "MS-" + s[i] + str(j+1) + "-" + "sell.csv"
            s3 = "MS-" + s[i] + str(j+1) + "-" + "supply.csv"
            names.append([s1, s2, s3])
    return names
```

```python
def process(name):

    df_sale = pd.read_csv("/home/major/msu/6sem/prak/3/out/input/" + name[1], sep=',')
    df_i = pd.read_csv("/home/major/msu/6sem/prak/3/out/input/" + name[0], sep=',')
    df_supply = pd.read_csv("/home/major/msu/6sem/prak/3/out/input/" + name[2], sep=',')

    cur_ap = int(df_supply.apple[0])
    cur_pe = int(df_supply.pen[0])

    q = []
    q.append(["data", "apple" , "pen"])
    stolen = []
    stolen.append(["data", "apple" , "pen"])

    cur_pos_sale = 0
    cur_pos_supply = 1
    cur_pos_i = 0
	
    info = []
    apple_sold = 0
    pen_sold = 0
    apple_st = 0
    pen_st = 0

    for i in range(df_i.shape[0]):
           # print(i)
            days_in_m = int(df_i.date[i][8])*10 + int(df_i.date[i][9])

            for k in range(days_in_m):
                next_d = df_sale.date[cur_pos_sale]
                cur_d = df_sale.date[cur_pos_sale]

                while cur_d == next_d :
                    if df_sale.sku_num[cur_pos_sale][6] == 'a':
                        cur_ap -= 1
                        apple_sold += 1                         
                    elif df_sale.sku_num[cur_pos_sale][6] == 'p':
                        cur_pe -= 1
                        pen_sold += 1
                    cur_pos_sale += 1
                    if cur_pos_sale >= df_sale.shape[0]:
                       break
                    next_d = df_sale.date[cur_pos_sale]

                s1 = df_sale.date[cur_pos_sale-1]
                s2 = cur_ap
                s3 = cur_pe
                q.append([s1, s2, s3])

                if k == 14:
                    cur_ap += int(df_supply.apple[cur_pos_supply])
                    cur_pe += int(df_supply.pen[cur_pos_supply])
                    cur_pos_supply += 1

            p1 = df_i.date[cur_pos_i]
            p2 = cur_ap - int(df_i.apple[cur_pos_i])
            p3 = cur_pe - int(df_i.pen[cur_pos_i])
            stolen.append([p1, p2, p3])
            apple_st += p2
            pen_st += p3
            
            if((i+1) % 12 == 0):
                year = 2006 + int((i+1)/12) - 1
                info.append([df_sale.sku_num[3][3], year, apple_sold, apple_st, pen_sold, pen_st]) 				

            if cur_pos_supply >= df_supply.shape[0]:
                break

            cur_ap = int(df_i.apple[cur_pos_i]) + int(df_supply.apple[cur_pos_supply])
            cur_pe = int(df_i.pen[cur_pos_i]) + int(df_supply.pen[cur_pos_supply])
            cur_pos_supply += 1
            cur_pos_i += 1
    print(name[0][0] + name[0][1] + name[0][2] + name[0][3] + name[0][4], " calculated") 
    return q, stolen, info
```
